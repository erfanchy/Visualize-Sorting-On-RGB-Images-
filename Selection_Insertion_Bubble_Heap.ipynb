```python
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
import random

def heap_sort(reverse_list,img):
    buildHeap(reverse_list)
    for endidx in reversed(range(1,len(reverse_list))) :
        swap(0,endidx,reverse_list)
        siftdown(0,endidx-1,reverse_list) 
        reversed_image = [reverse[m] for m in reverse_list]
        reversed_image_copy= np.array(reversed_image).reshape(img.shape[0],img.shape[1],img.shape[2])
        plt.xlabel("Width")
        plt.ylabel("Height")
        plt.title("Heap Sort",fontweight='bold',fontsize = 16)
        plt.imshow(reversed_image_copy)
        display.clear_output(wait=True)
        plt.show()
    return reverse_list 


def buildHeap(reverse_list) :

    firstparent = (len(reverse_list)-2)//2 
    for currentidx in reversed(range(firstparent+1)) : 
        siftdown(currentidx,len(reverse_list)-1,reverse_list)

def swap(i,j,reverse_list) : 

    reverse_list[i],reverse_list[j] = reverse_list[j],reverse_list[i]


def siftdown(currentidx,endidx,reverse_list) :

    childone = (currentidx * 2) + 1

    while childone <=endidx : 

        childtwo = (currentidx *2) + 2 if (currentidx*2+2)<=endidx else -1 

        if childtwo!=-1 and reverse_list[childone] < reverse_list[childtwo] : 

            idxtoswap = childtwo

        else : 
            idxtoswap = childone

        if reverse_list[currentidx] < reverse_list[idxtoswap] : 

            swap(currentidx,idxtoswap,reverse_list)
            currentidx = idxtoswap
            childone = (currentidx*2 +1) 
        else : 
            break 

def selection_sort(reverse_list,img) :
    for current in range(len(reverse_list)-1) :
        smallest = current 
        for i in range(current + 1 , len(reverse_list)) : 
            
            if reverse_list[smallest] > reverse_list[i] : 
                smallest = i
        reversed_image = [reverse[m] for m in reverse_list]
        reversed_image_copy= np.array(reversed_image).reshape(img.shape[0],img.shape[1],img.shape[2])
        plt.xlabel("Width")
        plt.ylabel("Height")
        plt.title("Selection Sort",fontweight='bold',fontsize = 16)
        plt.imshow(reversed_image_copy)
        display.clear_output(wait=True)
        plt.show()
        swap(current,smallest,reverse_list)
    return reverse_list
        
def bubble_sort(reverse_list,img):
    issorted = False 
    while not issorted  :
        issorted = True 
        for i in range(len(reverse_list)-1) : 
            if reverse_list[i] > reverse_list[i+1] :
                swap(i,i+1,reverse_list)
                issorted = False 
        reversed_image = [reverse[m] for m in reverse_list]
        reversed_image_copy= np.array(reversed_image).reshape(img.shape[0],img.shape[1],img.shape[2])
        plt.xlabel("Width")
        plt.ylabel("Height")
        plt.title("Bubble Sort",fontweight='bold',fontsize = 16)
        plt.imshow(reversed_image_copy)
        display.clear_output(wait=True)
        plt.show() 
    return reverse_list
    
def insertion_sort(reverse_list,img) :
    for i in range(1,len(reverse_list)) :
        j = i 
        while j>0 and reverse_list[j] < reverse_list[j-1] :
            swap(j,j-1,reverse_list)
            j-=1
        reversed_image = [reverse[m] for m in reverse_list]
        reversed_image_copy= np.array(reversed_image).reshape(img.shape[0],img.shape[1],img.shape[2])
        plt.xlabel("Width")
        plt.ylabel("Height")
        plt.title("Insertion Sort",fontweight='bold',fontsize = 16)
        plt.imshow(reversed_image_copy)
        display.clear_output(wait=True)
        plt.show()     
    return reverse_list

def swap(i,j,reverse_list) :
    
    reverse_list[i],reverse_list[j]= reverse_list[j],reverse_list[i]
    
if __name__=="__main__" :

    img= plt.imread("/home/erfan/Desktop/Walter_White_Sorting_Visualizer/walter_white.png")
    plt.imshow(img)
    plt.show()
    
    flattened_image = img.flatten()
    image = {} 
    for i in range(len(flattened_image)) : 
        image[i] = flattened_image[i]
    reverse = {}
    for key, value in (list(image.items())):
        reverse[key] =  value
    reverse_list = list(reverse.keys())
    random.Random(4).shuffle(reverse_list)
    answer = selection_sort(reverse_list,img)
    
```
